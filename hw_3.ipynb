```python
import sympy as sym
from sympy import Symbol
from sympy import pprint
```

```python
a=Symbol('a')
b=Symbol('b')
x=Symbol('x')
```

```python
my_f=(x-a)/(b-a)
sym.plot(my_f.subs({a:0,b:5}),(x,0,5),title='uniform distribution')
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
x_values=[]
y_values=[]
for value in range(0,5):
    y=my_f.subs({a:0,b:5,x:value}).evalf()
    y_values.append(y)
    x_values.append(value)
    print(value,y)
plt.plot(x_values,y_values)
plt.show()
```
